```python
import matplotlib.pyplot as plt

temperatures = [8.2, 17.4, 14.1, 7.9, 18.0, 13.5, 9.0, 17.8, 13.0, 8.5,
 16.5, 12.9, 7.7, 17.2, 13.3, 8.4, 16.7, 14.0, 9.5, 18.3, 13.4, 8.1,
 17.9, 14.2, 7.6, 17.0, 12.8, 8.0, 16.8, 13.7, 7.8, 17.5, 13.6, 8.7,
 17.1, 13.8, 9.2, 18.1, 13.9, 8.3, 16.4, 12.7, 8.9, 18.2, 13.1, 7.8,
 16.6, 12.5]

night = []
evening = []
day = []

for i, temp in enumerate(temperatures):
    if 0 <= i < 8:  # Night
        night.append(temp)
    elif 8 <= i < 16:  # Evening
        evening.append(temp)
    elif 16 <= i < 24:  # Day
        day.append(temp)

average_day_temperature = sum(day) / len(day) if day else 0

print("Night temperatures:", night)
print("Evening temperatures:", evening)
print("Day temperatures:", day)
print("Average day-time temperature:", average_day_temperature)

plt.plot(range(len(day)), day, marker='o', label='Day Temperatures')
plt.title("Day vs. Temperature")
plt.xlabel("Hour (16-24)")
plt.ylabel("Temperature (C)")
plt.legend()
plt.grid()
plt.show()
```

```python
def sum_nested_list(nested_list):
    """
    Recursively sums all numbers in a nested list.
    
    :param nested_list: A list that contains numbers and/or other nested lists of numbers.
    :return: Total sum of all numbers in the structure.
    """
    total_sum = 0
    for item in nested_list:
        if isinstance(item, list):  # If the item is a list, call the function recursively
            total_sum += sum_nested_list(item)
        else:  # Otherwise, add the number directly
            total_sum += item
    return total_sum

# Test the function with a sample nested list
nested_list = [1, [2, [3, 4], 5], 6, [7, 8]]
result = sum_nested_list(nested_list)

print("Total sum of all numbers in the nested list:", result)
```

```python
def generate_permutations(s):
    """
    Recursively generates all unique permutations of a string.

    :param s: The input string.
    :return: A list of unique permutations.
    """
    # Base case: If the string has a single character, return it as a list
    if len(s) <= 1:
        return [s]

    # Recursive case
    permutations = []
    for i, char in enumerate(s):
        # Remaining characters after removing the current one
        remaining = s[:i] + s[i+1:]

        # Generate permutations of the remaining characters
        for perm in generate_permutations(remaining):
            # Form a new permutation by prepending the current character
            new_perm = char + perm
            if new_perm not in permutations:  # Ensure uniqueness
                permutations.append(new_perm)

    return permutations

# Test the function
print("Permutations of 'abc':", generate_permutations("abc"))
print("Permutations of 'aab':", generate_permutations("aab"))
```

```python
def calculate_directory_size(directory):
    """
    Recursively calculates the total size of a directory, including nested files and subdirectories.

    :param directory: A dictionary representing the directory structure. 
                      Keys are names of files/directories, and values are either:
                      - An integer (file size in KB)
                      - Another dictionary (subdirectory)
    :return: Total size of the directory in KB.
    """
    total_size = 0

    for name, content in directory.items():
        if isinstance(content, int):  # If it's a file, add its size
            total_size += content
        elif isinstance(content, dict):  # If it's a subdirectory, recurse
            total_size += calculate_directory_size(content)

    return total_size

# Example directory structure
filesystem = {
    "file1.txt": 10,
    "file2.txt": 20,
    "subdir1": {
        "file3.txt": 15,
        "subsubdir1": {
            "file4.txt": 5,
            "file5.txt": 10
        }
    },
    "subdir2": {
        "file6.txt": 25
    }
}

# Calculate and print the total size of the root directory
total_size = calculate_directory_size(filesystem)
print(f"Total size of the directory: {total_size} KB")
```

```python
def min_coins(coins, amount):
    """
    Finds the minimum number of coins needed to make up the given amount.
    
    :param coins: List of coin denominations.
    :param amount: Target amount.
    :return: Minimum number of coins needed, or -1 if not possible.
    """
    # Create a DP array initialized with a large value (amount + 1, a value greater than any valid answer)
    dp = [float('inf')] * (amount + 1)
    
    # Base case: 0 coins are needed to make the amount 0
    dp[0] = 0

    # Build the DP table
    for coin in coins:
        for x in range(coin, amount + 1):
            dp[x] = min(dp[x], dp[x - coin] + 1)

    # If dp[amount] is still `inf`, it means the amount cannot be made with the given coins
    return dp[amount] if dp[amount] != float('inf') else -1

# Test the function
coins = [1, 2, 5]
amount = 11
result = min_coins(coins, amount)
print(f"Minimum coins needed to make {amount}: {result}")
```

```python
def longest_common_subsequence(s1, s2):
    """
    Finds the length of the longest common subsequence (LCS) of two strings using DP.
    
    :param s1: First string.
    :param s2: Second string.
    :return: Length of the LCS.
    """
    # Get lengths of the input strings
    m, n = len(s1), len(s2)
    
    # Create a DP table with dimensions (m+1) x (n+1), initialized to 0
    dp = [[0] * (n + 1) for _ in range(m + 1)]

    # Fill the DP table
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if s1[i - 1] == s2[j - 1]:  # Characters match
                dp[i][j] = dp[i - 1][j - 1] + 1
            else:  # Take the maximum of excluding either character
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])

    # The value at dp[m][n] is the length of the LCS
    return dp[m][n]

# Test the function
s1 = "abcde"
s2 = "ace"
result = longest_common_subsequence(s1, s2)
print(f"Length of LCS between '{s1}' and '{s2}': {result}")
```

```python
def knapsack(weights, values, capacity):
    """
    Solves the 0/1 Knapsack Problem using Dynamic Programming.

    :param weights: List of item weights.
    :param values: List of item values.
    :param capacity: Maximum weight capacity of the knapsack.
    :return: Maximum value that can be achieved within the given capacity.
    """
    n = len(weights)  # Number of items

    # Create a DP table with dimensions (n+1) x (capacity+1), initialized to 0
    dp = [[0] * (capacity + 1) for _ in range(n + 1)]

    # Fill the DP table
    for i in range(1, n + 1):  # Iterate over items
        for w in range(1, capacity + 1):  # Iterate over capacities
            if weights[i - 1] <= w:  # Current item can fit in the knapsack
                dp[i][w] = max(dp[i - 1][w], dp[i - 1][w - weights[i - 1]] + values[i - 1])
            else:  # Current item cannot fit in the knapsack
                dp[i][w] = dp[i - 1][w]

    # The maximum value is stored in dp[n][capacity]
    return dp[n][capacity]

# Test the function
weights = [1, 3, 4, 5]
values = [1, 4, 5, 7]
capacity = 7
result = knapsack(weights, values, capacity)
print(f"Maximum value in knapsack with capacity {capacity}: {result}")
```
